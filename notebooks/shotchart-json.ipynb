```python
#!/usr/bin/python
# import modules, connect to database
import json
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
JSONPATH = '/home/gmf/Code/repos/nba/json'
gameid = '0041400101'
json_file = JSONPATH + '/sc_' + gameid + '.json'
shots = json.loads(open(json_file, 'r').read())['rowSet']

def jet(N):
  C = np.zeros([N,3])
  M = np.floor(N/4)
  dN = 1/M
  C[0,2] = (M/2+1)/M
  i=1
  for j in range(1,int(M/2)):
    C[i,:] = C[i-1,:]
    C[i,2] += dN
    i+=1
  for j in range(0, int(M)):
    C[i,:] = C[i-1,:]
    C[i,1] += dN
    i+=1
  for j in range(0, int(M)):
    C[i,:] = C[i-1,:]
    C[i,0] += dN
    C[i,2] -= dN
    i+=1
  for j in range(0, int(M)):
    C[i,:] = C[i-1,:]
    C[i,1] -= dN
    i+=1
  for j in range(0,int(M/2)):
    C[i,:] = C[i-1,:]
    C[i,0] -= dN
    i+=1
  return C

```

```python
# scatter plot of missed and made shots

# draw court
minx = -250
maxx = 250
miny = -55
maxy = 845
img = mpimg.imread('/home/gmf/Code/repos/nba/nba_court.png')
implot = plt.imshow(img,extent=[minx, maxx, miny, maxy], zorder=0)

for s in shots:
  x,y,made = s[17],s[18],s[20]
  if made: # made shot
    plt.plot(x,y,'bo',zorder=1)
  else: # missed shot
    plt.plot(x,y,'rx',zorder=1)
plt.show()
```

```python
# heat map of FG%
#
 
# bin space
dx = 100
dy = 100
minx = -250
maxx = 250
miny = -55
maxy = 845
xbins = range(minx, maxx+1, dx)
ybins = range(miny, maxy+1, dy)
xcent = range(minx+dx/2, maxx+1, dx)
ycent = range(miny+dy/2, maxy+1, dy)
Rx = np.array([-1, -1, 1, 1])*dx/2
Ry = np.array([-1, 1, 1, -1])*dy/2

FGA = 0*np.meshgrid(np.array(xbins),np.array(ybins))[0]
FGM = 0*np.meshgrid(np.array(xbins),np.array(ybins))[0]
C = np.meshgrid(np.array(xcent),np.array(ycent))[0]
N = 64
colors = jet(N)
#colors = colormap(N)

for s in shots:
  x,y,made = s[17],s[18],s[20]
  j = (x - xbins[0])/dx
  i = (y - ybins[0])/dy
  FGA[i,j]+=1
  if made: FGM[i,j]+=1

# plot
# draw court
img = mpimg.imread('temp/nba_court.png')
implot = plt.imshow(img,extent=[minx, maxx, miny, maxy], zorder=0)  
# draw heat map        
FGP = float(0)*FGM
for i in range(0,np.size(C,0)):
  for j in range(0,np.size(C,1)):
    if FGA[i,j]>0:
      FGP[i,j] = float(FGM[i,j])/FGA[i,j]
    else:
      FGP[i,j] = 0
    fgi = np.round(FGP[i,j]*(N-1))
    plt.fill(xcent[j]+Rx,ycent[i]+Ry,color=colors[fgi,:], zorder=1, alpha=0.7)
plt.show()
```
